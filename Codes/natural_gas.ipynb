# Natural Gas Spot Prices 
Henry Hub Benchmark: Henry Hub is the primary U.S. benchmark price for natural gas, and EIA publishes daily and weekly spot prices for it. Volatility in Henry Hub quickly translates to power price volatility, since gas-fired generators set prices in ERCOT during many hours. For example, Henry Hub spiked in January 2024 during a cold snap (nearly \$10/MMBtu) and then crashed by late February​, illustrating how weather-driven gas price swings can lead to sudden jumps and drops in electricity prices.
Texas Regional Hubs: In ERCOT, gas price dynamics at Texas hubs like the Houston Ship Channel (HSC) in East Texas and Waha Hub in West Texas are especially important. These hubs often trade at a “basis” differential to Henry Hub. EIA data show that in the first half of 2023, with improved supply conditions, HSC averaged about \$0.27 per MMBtu below Henry Hub and Waha about \$0.85 below. In practice, when Texas spot gas is cheap, ERCOT power prices tend to be lower (all else equal), whereas if Texas gas trades at a premium (e.g. during a freeze or pipeline outage), ERCOT power prices jump accordingly. Virtual traders watch these hub prices closely; any rapid change in Henry Hub or Texas basis can signal an upcoming divergence between expected DA prices and real-time outcomes.

## Natural Gas Storage

Natural gas storage acts as a buffer for the market, and storage data from EIA’s Weekly Natural Gas Storage Report provide crucial supply insight. Texas is part of the South Central storage region, which also includes nearby Gulf states – a key supply source for Texas gas needs. When storage levels in this region are below normal, it signals that less gas is in reserve for peak demand, often leading to higher gas price volatility in Texas. As of late February 2025, U.S. storage was drawn down to 1,840 Bcf, about 11% below the five-year average for that date​. Such deficits in storage put upward pressure on gas prices. In ERCOT’s context, if heading into summer or winter the South Central storage is low, market participants expect higher gas (and power) prices during heat waves or cold snaps because there’s less cushion to meet spiking demand. Conversely, high storage inventories (above average) generally dampen gas price rallies, which can keep a lid on power prices.

```python
tx_gas_storage = pd.read_excel('NG_STOR_WKLY_S1_W.xls', sheet_name = 'Data 1', skiprows=2)
tx_gas_storage.tail()
```

## Production

Texas is the nation’s top gas producer, accounting for roughly 27% of U.S. natural gas production in 2023​. EIA data on production (from sources like the Drilling Productivity Report and state production reports) show how trends in major basins can shift the supply picture for ERCOT. Two key supply areas are the Permian Basin (West Texas, primarily associated gas from oil drilling) and the Eagle Ford Shale (South Texas). Rapid growth in these regions in recent years has greatly increased gas availability in Texas. In 2022 alone, Permian natural gas production jumped 15% (+2.7 Bcf/d) to a record 21.2 Bcf/d, and Eagle Ford output rose about 14% (+0.8 Bcf/d)​. This surge (enabled by advances in horizontal drilling and fracturing) pushed Texas’s total gas withdrawals to an all-time high of 12.4 Tcf in 2023​.

## Henry Hub and Futures 

In practice, forward gas curves and EIA forecasts influence ERCOT market behavior by shaping expectations. Power market participants often incorporate gas futures into their bids (e.g. a generator locking in fuel for future months or a load hedging costs). Virtual traders, who profit from price differences between DA and RT, use these forward indicators to spot mispricings – for instance, if the day-ahead market underestimates a looming heat wave’s impact on gas (and thus power) prices, a virtual trader armed with EIA’s weather-driven demand insight might take a long position in DA, betting real-time prices will spike. Conversely, if forward data suggests a supply glut is coming, a trader might sell in DA expecting real-time prices to slump.

```python
tx_gas_henry = pd.read_excel('NG_PRI_FUT_S1_D.xls', sheet_name = 'Data 1', skiprows=2)
tx_gas_henry.tail()
```
